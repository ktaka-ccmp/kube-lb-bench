```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df02=pd.read_csv("cpu02.csv")
df04=pd.read_csv("cpu04.csv")
df08=pd.read_csv("cpu08.csv")
df12=pd.read_csv("cpu12.csv")
df16=pd.read_csv("cpu16.csv")

print(plt.style.available)
```

```python
plt.style.use('seaborn-poster')
ax=df02.plot.scatter(x='# of pods',y='Req/sec', color='m', label='2cpu')
ax=df04.plot.scatter(x='# of pods',y='Req/sec', color='r', label='4cpu', ax=ax)
ax=df08.plot.scatter(x='# of pods',y='Req/sec', color='b', label='8cpu', ax=ax)
ax=df12.plot.scatter(x='# of pods',y='Req/sec', color='g', label='12cpu', ax=ax)
ax=df16.plot.scatter(x='# of pods',y='Req/sec', color='c', label='16cpu', ax=ax)
ax.set_xlim(0,50)
ax.set_ylim(0,120000)
plt.savefig('Repeatability3.png')
```
